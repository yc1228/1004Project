```python
import pandas as pd
import time
from pandas import DataFrame
%matplotlib inline 
import matplotlib.pyplot as plt
df_cmdt = DataFrame.from_csv("/Users/zoem/Documents/DS1004/term_project/result/clo_2_result.out",header=None,index_col=False,sep="\t")

df_cmdt.columns = ["Date","Year","Month","Day","First_type","Second_type","Data_property"]
df_cmdt=df_cmdt.fillna("Null")
df_cmdt.head()
```

```python
df_cmdt.loc[df_cmdt['Data_property']=='Valid'].groupby('Year').count()
```

```python
ct_year=df_cmdt.loc[df_cmdt['Data_property']=='Valid'].groupby('Year').count()
fig, ax = plt.subplots(figsize=(10, 5))
import numpy as np 
year=np.arange(2006,2016)
count = np.array(ct_year['Date'])[1:11]

ax.barh(year, count, align='center',
        color='indianred')
ax.set_yticks(year)
ax.set_yticklabels(year)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Count')
ax.set_title('Complaint every year based on start date')

plt.show()
```

```python
ct_month=df_cmdt.loc[df_cmdt['Data_property']=='Valid'].groupby('Month').count()
ct_month
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
import numpy as np 
month=np.arange(1,13)
count = np.array(ct_month['Date'])
ax.barh(month, count, align='center',
        color='mistyrose')
ax.set_yticks(month)
ax.set_yticklabels(month)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Count')
ax.set_title('Complaint every month based on start date')

plt.show()
```

```python
ct_day=df_cmdt.loc[df_cmdt['Data_property']=='Valid'].groupby('Day').count()
fig, ax = plt.subplots(figsize=(10, 8))
import numpy as np 
day=np.arange(1,32)
count = np.array(ct_day['Date'])
ax.barh(day, count, align='center',
        color='salmon')
ax.set_yticks(day)
ax.set_yticklabels(day)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Count')
ax.set_title('Complaint every month based on start date')

plt.show()
```
